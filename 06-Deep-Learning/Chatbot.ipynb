```python
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(filters=[])
tokenizer.fit_on_texts(vocab)
```

```python
tokenizer.word_index
```

```python
train_story_text = []
train_question_text = []
train_answers = []
```

```python
for story,question,answer in train_data:
    train_story_text.append(story)
    train_question_text.append(question)
    train_answers.append(answer)
```

```python
def vectorize_stories(data, word_index=tokenizer.word_index, max_story_len=max_story_len, max_question_len=max_question_len):
    
    # STORIES = X
    X = []
    # QUESTIONS Xq
    Xq = []
    Y =[]
    
    for story, query, answer in data:
        
        # vector for each story
        x = [word_index[word.lower()] for word in story]
        xq = [word_index[word.lower()] for word in query]
        y = np.zeros(len(word_index) + 1)
        y[word_index[answer]] = 1
        
        X.append(x)
        Xq.append(xq)
        Y.append(y)
        
    return (pad_sequences(X, maxlen=max_story_len), pad_sequences(Xq,maxlen=max_question_len), np.array(Y))
```

```python
inputs_train, queries_train, answers_train = vectorize_stories(train_data)
```

```python
input_test, queries_test, answers_test = vectorize_stories(test_data)
```

```python
from keras.models import Sequential,Model
from keras.layers.embeddings import Embedding
from keras.layers import Input,Activation,Dense,Permute,Dropout,add,dot,concatenate,LSTM
```

```python
# PLACEHOLDER shape=(max_story_len, batch_size)
input_sequence = Input((max_story_len, ))
question = Input((max_question_len, ))
```

```python
# vocab_len
vocab_size = len(vocab) + 1
```

```python
# INPUT ENCODER M
input_encoder_m = Sequential()
input_encoder_m.add(Embedding(input_dim=vocab_size, output_dim=64))
input_encoder_m.add(Dropout(0.3))

# OUTPUT
# (samples, story_maxlen, embedding_dim)
```

```python
# INPUT ENCODER C
input_encoder_c = Sequential()
input_encoder_c.add(Embedding(input_dim=vocab_size, output_dim=max_question_len))
input_encoder_c.add(Dropout(0.3))

# OUTPUT
# (samples, story_maxlen, max_question_len)
```

```python
question_encoder = Sequential()
question_encoder.add(Embedding(input_dim=vocab_size, output_dim=64, input_length=max_question_len))
question_encoder.add(Dropout(0.3))

# OUTPUT
# (samples, query_maxlen, embedding_dim)
```

```python
input_encoded_m = input_encoder_m(input_sequence)
input_encoded_c = input_encoder_c(input_sequence)
question_encoded = question_encoder(question)
```

```python
match = dot([input_encoded_m, question_encoded], axes=(2,2))
match = Activation('softmax')(match)
```

```python
response = add([match, input_encoded_c])
response = Permute((2, 1))(response)
```

```python
answer = concatenate([response, question_encoded])
```

```python
answer
```

```python
answer = LSTM(32)(answer)
```

answer = Dropout(0.5)

```python
answer = Dropout(0.5)(answer)
```

```python
answer = Dense(vocab_size)(answer)
```

```python
answer = Activation('softmax')(answer)
```

```python
model = Model([input_sequence, question],answer)
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

model.summary()

```python
model.summary()
```

```python
history = model.fit([inputs_train, queries_train], answers_train,batch_size=32,epochs=4,validation_data=([input_test,queries_test],answers_test))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
